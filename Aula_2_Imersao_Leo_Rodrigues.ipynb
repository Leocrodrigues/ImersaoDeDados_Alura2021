#Desafio 02: Melhorar a visualização alterando tamanho da fonte...

#Desafio 04: Estudar sobre as estatíticas retornadas no .describe()

- Count = Quantidade(total) da amostra
- mean: É média aritmética simples para a variável, ou seja, é a soma dos valores (Xi)
- std: O desvio padrão é a raiz quadrada da variância, sendo ela uma medida derivada da variância, tornando mais simples a interpretação
da dispersão em torno da média
- Min: Valor minimo das amostras de cada classe
- 1 Quartil: 25% das observações são menores do que o 1º quartil
- 2 Quartil: trata-se da mediana
- 3 Quartil: 25% das observações são maiores do que o 3º quartil
- Max: Valor maximo das amostras de cada classe

#Desafio 06: Fazer outras análises com o boxplot e até com o histograma. 

```python
plt.figure(figsize=(10,8))
sns.catplot(y='c99', x='tempo', data=dados, hue='dose', col='tratamento', kind="box")
```

Podemos ver nesse boxplot que, com tratamento 'com droga' temos valores resultantes distante das observações. Já, o com controle nao, os dados estao proximo indicando que os dados nao variam tanto e os valores resultantes estao proximos

```python
tratamentos = dados['tratamento'].unique()
dados = pd.DataFrame({t : dados.query('tratamento == @t').loc[:,'g0':'g771'].describe().T['mean'] for t in tratamentos})
```

```python
plt.figure(figsize=(10,8))
ax = sns.histplot(data=dados, kde=True)
ax.set_xlabel('Médias dos genes (0 a 771)')
ax.set_ylabel('Frequência')
plt.show()
```

Podemos verificar que a média do tratamento 'controle' e menor e simetrica. Já, o tratamento 'com droga' tambem e simetrica, porem com uma distribuiçao mais alongada.
