#Desafio 06: Fazer outras anÃ¡lises com o boxplot e atÃ© com o histograma. 

```python
plt.figure(figsize=(10,8))
sns.catplot(y='c99', x='tempo', data=dados, hue='dose', col='tratamento', kind="box")
```

Podemos ver nesse boxplot que, com tratamento 'com droga' temos valores resultantes distante das observaÃ§Ãµes. JÃ¡, o com controle nao, os dados estao proximo indicando que os dados nao variam tanto e os valores resultantes estao proximos

```python
tratamentos = dados['tratamento'].unique()
dados = pd.DataFrame({t : dados.query('tratamento == @t').loc[:,'g0':'g771'].describe().T['mean'] for t in tratamentos})
```

```python
plt.figure(figsize=(10,8))
ax = sns.histplot(data=dados, kde=True)
ax.set_xlabel('MÃ©dias dos genes (0 a 771)')
ax.set_ylabel('FrequÃªncia')
plt.show()
```

Podemos verificar que a mÃ©dia do tratamento 'controle' e menor e simetrica. JÃ¡, o tratamento 'com droga' tambem e simetrica, porem com uma distribuiÃ§ao mais alongada.
